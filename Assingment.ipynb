```python
### QUESTIION NO 13

class annual_fee():
  def __init__(self,name):
    self.name=name
    self.anual_cost=[]

  def subject(self):
      self.subject=input("Enter which subject you want to take for masters (HR,Finance,DS,Marketing)")
      if self.subject=='DS':
          print("there is not available analytics course for this core subject DS")
      else:
          print(f"{self.name} is takking  {self.subject} subject")

  def ADD_EXPENSES_Course(self):
    self.hostel=input("is your course anlytics")
    if self.hostel == "yes":
      self.anual_cost.append(200000+200000*0.1)
      print(self.anual_cost)
    else:
      return "Thnak you !"

  def ADD_EXPENSES_Hostel(self):
    self.hostel=input("Do you want collage hostel facilities")
    if self.hostel=="yes":
        self.anual_cost.append(200000)
    else:
        print("Thank You !")

  def ADD_EXPENSES_food(self):
    self.Months=int(input("Enter how many moths you want food"))
    self.anual_cost.append(self.Months*2000)
    return  self.anual_cost

  def ADD_EXPENSES_transportation(self):
    self.transportation=int(input("How many semester you want Transpotations services"))
    self.anual_cost.append(self.transportation*13000)
    return  self.anual_cost

  def TOTAL_ANNUAL_EXPENCES(self):
    total=0
    for i in self.anual_cost:
      total+=i
    return f" {self.name} Your total Anual cost for this {self.subject} subject : {total}₹"

```

```python
## QUESTION No 14
# Price data
store = {
    "books": {
        1: {"hindi": 60, "marathi": 60, "english": 80, "science": 90, "math": 100},
        2: {"hindi": 60, "marathi": 60, "english": 80, "science": 90, "math": 100},
        3: {"hindi": 60, "marathi": 60, "english": 80, "science": 90, "math": 100},
        4: {"hindi": 60, "marathi": 60, "english": 80, "science": 90, "math": 100},
        5: {"hindi": 100, "marathi": 100, "english": 100, "science": 120, "math": 140},
        6: {"hindi": 100, "marathi": 100, "english": 100, "science": 120, "math": 140},
        7: {"hindi": 100, "marathi": 100, "english": 100, "science": 120, "math": 140},
        8: {"hindi": 100, "marathi": 100, "english": 100, "science": 120, "math": 140},
        9: {"hindi": 150, "marathi": 150, "english": 150, "science": 200, "math": 250},
        10: {"hindi": 150, "marathi": 150, "english": 150, "science": 200, "math": 250}
    },

    "notebooks": {
        100: {"square": 40, "4lines": 30, "2lines": 30, "single lines": 60, "A4 notebook": 100},
        200: {"square": 70, "4lines": 50, "2lines": 50, "single lines": 100, "A4 notebook": 180}
    }
}

cart = {}
total_amount = 0

### FirstFunction for Take customer order and calculate bill
def purchase_items():
    global total_amount
    while True:
        category =input("What do you want (books / notebooks): ").lower()

        if category == "books":
            print(store["books"].keys())

            std =int(input("Enter standard: "))

            if std in store["books"]:
              
                print(store["books"][std].keys())
                item = input("Enter subject: ")
                qty = int(input("Quantity: "))

                if item in store["books"][std]:

                    price = store["books"][std][item]
                    item_total = price * qty
                    total_amount += item_total
                    cart[item] = (qty, item_total)

        elif category == "notebooks":
            print(store["notebooks"].keys())
            pages = int(input("Enter pages: "))

            if pages in store["notebooks"]:
                print( store["notebooks"][pages].keys())
                item = input("Enter notebook type: ")
                qty = int(input("Quantity: "))

                if item in store["notebooks"][pages]:
                    price = store["notebooks"][pages][item]
                    item_total = price * qty
                    total_amount += item_total
                    cart[item] = (qty, item_total)

        more = input("Do you want to buy more (yes/no): ").lower()
        if more != "yes":
            break


def print_bill():
    """Print final bill"""
    line = "_"*50
    print(line)
    print("{:^50}".format("Ambika Stationery"))
    print(line)
    print("| {:10} {:>10} {:>10} |".format("Product", "Qty", "Price"))
    print(line)

    for item in cart:
        print("| {:10} {:>10} {:>10} |".format(item, cart[item][0], cart[item][1]))

    print(line)
    print("| {:10} {:>22} |".format("Total", total_amount))
    print(line)


def start_shopping():
    """Main function"""
    choice =input("Do you want to purchase (yes/no): ").lower()

    if choice == "yes":
        purchase_items()
        bill =input("Print bill (yes/no): ").lower()
        if bill == "yes":
            print_bill()
    else:
        print("Thank you")


# Program start
start_shopping()

```

```python
## QUESTION NO 44

import pyodbc

# SQL Server connection
connection = pyodbc.connect(
    r'Driver={SQL Server};'
    r'Server=KARAN\SQLEXPRESS;'
    r'Database=Restourant;'
    r'Trusted_Connection=yes;'
)
cursor = connection.cursor()

# Create table (only once)
cursor.execute("""
IF NOT EXISTS (SELECT * FROM sys.tables WHERE name = 'cust_info')
CREATE TABLE cust_info (
    sr_no INT,
    name VARCHAR(50),
    dob VARCHAR(20),
    mobile VARCHAR(10))
""")
connection.commit()

# Take input
sr_no = input("Enter sr no: ")
name = input("Enter name: ")
dob = input("Enter DOB (YYYY-MM-DD): ")
mobile = input("Enter mobile: ")

# SIMPLE validation
if sr_no == "" or name == "" or dob == "" or mobile == "":
    print("All fields are required")
    exit()    ### (EXIT) IT STOP EXICUTION OF PROGRAM WHEN INPUT IS NOT VALID 

if not sr_no.isdigit():
    print("sr_no must be number")
    exit()

if not mobile.isdigit() or len(mobile) != 10:
    print("Mobile must be 10 digits")
    exit()

# Insert data
cursor.execute(
    "INSERT INTO cust_info VALUES (?,?,?,?)",
    int(sr_no), name, dob, mobile
)

connection.commit()  ## USING THIS WE COMMIT DATA AND SAVE DATA IN DATABASE PARMANTLY 
print("Data inserted successfully")

cursor.close()
connection.close()

```
