```python
# 1. Square of number from 1 to 10
squaree = [i*i for i in range(1, 11)]
print(squaree)
```

```python
# 2. Even numbers between 1 and 50
even= [i for i in range(1, 51) if i % 2 == 0]
print(even)
```

```python
# 3 Convert strings to uppercase
names = ["karan", "rajesh", "shende"]
up = [i.upper() for i in names]
print(up)
```

```python
# 4. Only positive numbers
nums = [-3, 5, -1, 7, 0, 9]
pos = [i for i in nums if i > 0]
print(pos)
```

```python
# 5.Tuples (num, num^2) from 1 to 5
tupp = [(i, i*i) for i in range(1, 6)]
print(tupp)
```

```python
# 6. Extract vowels from string
s = "education"
v = [i for i in s if i in "aeiou"]
print(v)
```

```python
# 7.Flatten 2D list
lst = [[1, 2], [3, 4], [5, 6]]
flat = [j for i in lst for j in i]
print(flat)
```

```python
# 8. Replace negative numbers with 0
nums = [-2, 4, -5, 6]
new = [0 if i < 0 else i for i in nums]
print(new)
```

```python
# 9. Length of each word
words = ["python", "java", "c"]
lengths = [len(i) for i in words]
print(lengths)
```

```python
# 10. Words not starting with A or a
words = ["Apple", "ball", "ant", "cat"]
res = [i for i in words if not i.lower().startswith('a')]
print(res)
```

```python
# 11. Even or odd list
nums = [1, 2, 3, 4]
eo = ["even" if i % 2 == 0 else "odd" for i in nums]
print(eo)
```

```python
# 12. Numbers divisible by 3 and 5 (1–100)
d = [i for i in range(1, 101) if i % 3 == 0 and i % 5 == 0]
print(d)
```

```python
# 13. Multiplication table 1–5
mul = [[i*j for j in range(1, 6)] for i in range(1, 6)]
print(mul)
```

```python
# 14. Dictionary keys to list
d = {"a": 1, "b": 2}
KEYS = [i for i in d]
print(KEYS)
```

```python
# 15. Digits from string
DATA= "ab12cd34"
dig = [i for i in DATA if i.isdigit()]
print(dig)
```

```python
# 16. Remove spaces from string
S= "hello world"
nospace = [i for i in S if i != " "]
print("".join(nospace))
```

```python
# 17. Characters appearing more than once
s = "programming"
name = [i for i in s if s.count(i) > 1]
print(name)
```

```python
# 18. All words from sentences
sentance = ["hello world", "python is easy"]
word= [w for s in sentance for w in s.split()]
print(word)
```

```python
# 19. Unique elements from list
listt= [1, 2, 2, 3, 4, 4]
uniq = [i for i in listt if lst.count(i) == 1]
print(uniq)
```

```python
# 20. CARTITION PRODUCT
a = [1, 2]
b = [3, 4]
pairs = [(i, j) for i in a for j in b]
print(pairs)
```

```python
# 1. Add two numbers
add = lambda x, y: x + y
print(add(3, 4))
```

```python
# 2. Check even
iseven = lambda x: x % 2 == 0
print(iseven(6))
```

```python
# 3. Last character of string
last = lambda s: s[-1]
print(last("python"))
```

```python
# 4. Square numbers using map
Num = [1, 2, 3]
squaree= list(map(lambda x: x*x, Num))
print(squaree)
```

```python
# 5. Odd numbers using filter
oddd = list(filter(lambda x: x % 2 != 0, nums))
print(oddd)
```

```python
# 6. Sort tuples by second value
t = [(1, 3), (4, 1), (2, 2)]
print(sorted(t, key=lambda x: x[1]))
```

```python
# 7. Palindrome check
pal = lambda s: s == s[::-1]
print(pal("madam"))
```

```python
# 8. Max of three numbers
mxxx = lambda a, b, c: a if a > b and a > c else b if b > c else c
print(mxxx(3, 9, 5))
```

```python
# 9. Reverse string
reve = lambda s: s[::-1]
print(reve("hello"))
```

```python
# 10. String to int using map
NUMS = list(map(lambda x: int(x), ["1", "2", "3"]))
print(NUMS)
```

```python
# 11. Remove empty strings
lst = ["hi", "", "bye", ""]
clean = list(filter(lambda x: x != "", lst))
print(clean)
```

```python
# 13. Larger of two numbers
LARGE = lambda a, b: a if a > b else b
print(LARGE(4, 9))
```

```python
# 14. Divisible by 5
dIV_5= lambda x: x % 5 == 0
print(dIV_5(20))
```

```python
# 15. Add 10 to each element
nums = [1, 2, 3]
plus10 = list(map(lambda x: x + 10, nums))
print(plus10)
```

```python
# 16. Sort list of dict by age
data = [{"age": 25}, {"age": 20}]
print(sorted(data, key=lambda x: x["age"]))
```

```python
# 17. Check vowel
vowel = lambda c: c.lower() in "aeiou"
print(vowel("a"))
```

```python
# 18. WORD LENGTH GREATER THAN 5
words = ["python", "java", "programming"]
long = list(filter(lambda x: len(x) > 5, words))
print(long)
```

```python
# 19. Area of circle
import math
from math import *
area = lambda r: math.pi * r * r
print(area(5))
```

```python
# 20. Remove duplicates
lst = [1, 2, 2, 3]
nodup = list(filter(lambda x: lst.count(x) == 1, lst))
print(nodup)
```

```python
# 24. Uppercase characters from string
s = "PyThOn"
up = list(filter(lambda x: x.isupper(), s))
print(up)
```

```python
# 27. Anagram check
anagr= lambda a, b: sorted(a) == sorted(b)
print(anagr("listen", "silent"))
```

```python
# 28. Numeric values from mixed list
mix = [1, "a", 3, "b"]
num = list(filter(lambda x: type(x) == int, mix))
print(num)
```

```python
# 29. Any negative number
nums = [1, -2, 3]
print(any(map(lambda x: x < 0, nums)))
```
