```python
### QUESTIION NO 13

class annual_fee():
  def __init__(self,name):
    self.name=name
    self.anual_cost=[]

  def subject(self):
      self.subject=input("Enter which subject you want to take for masters (HR,Finance,DS,Marketing)")
      if self.subject=='DS':
          print("there is not available analytics course for this core subject DS")
      else:
          print(f"{self.name} is takking  {self.subject} subject")

  def ADD_EXPENSES_Course(self):
    self.hostel=input("is your course anlytics")
    if self.hostel == "yes":
      self.anual_cost.append(200000+200000*0.1)
      print(self.anual_cost)
    else:
      return "Thnak you "

  def ADD_EXPENSES_Hostel(self):
    self.hostel=input("Do you want collage hostel facilities")
    if self.hostel=="yes":
        self.anual_cost.append(200000)
    else:
        print("Thank You ")

  def ADD_EXPENSES_food(self):
    self.Months=int(input("Enter how many moths you want food"))
    self.anual_cost.append(self.Months*2000)
    return  self.anual_cost

  def ADD_EXPENSES_transportation(self):
    self.transportation=int(input("How many semester you want Transpotations services"))
    self.anual_cost.append(self.transportation*13000)
    return  self.anual_cost

  def TOTAL_ANNUAL_EXPENCES(self):
    total=0
    for i in self.anual_cost:
      total+=i
    return f" {self.name} Your total Anual cost for this {self.subject} subject : {total}₹"

```

```python
Karan=annual_fee("karan")
Karan.subject()
Karan.ADD_EXPENSES_Course()
Karan.ADD_EXPENSES_Hostel()
Karan.ADD_EXPENSES_food()
Karan.ADD_EXPENSES_transportation()
Karan.TOTAL_ANNUAL_EXPENCES()
```
