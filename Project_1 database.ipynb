```python
import pyodbc

# 1. Connect to SQL Server
connection = pyodbc.connect(
    r'Driver={SQL Server};'
    r'Server=KARAN\SQLEXPRESS;'
    r'Database=Restourant;'
    r'Trusted_Connection=yes;'
)

cursor =connection.cursor()

#2. Show Menu
cursor.execute("SELECT item_id, item_name, price FROM Menu")
menu_items = cursor.fetchall()

print("MENU")
for item in menu_items:
    print(f"{item.item_id}. {item.item_name} - ₹{item.price}")

# 3.Take Order
item_id=int(input("Enter Item ID: "))
quantity=int(input("Enter Quantity: "))

# 4.Fetch Price
cursor.execute("SELECT price FROM Menu WHERE item_id = ?", item_id)
price=cursor.fetchone()[0]

# 5.Calculate Total
total_price=price * quantity

# 6.Insert into Orders table
cursor.execute(
    """
    INSERT INTO Orderss (item_id, Quantity, Total_Price)
    VALUES (?, ?, ?)
    """,
    (item_id, quantity, total_price)
)

connection.commit()

print("Order Placed Successfully!")
print(f"Total Bill Amount: ₹{total_price}")

# 7.Close Connection
cursor.close()
connection.close()
```
