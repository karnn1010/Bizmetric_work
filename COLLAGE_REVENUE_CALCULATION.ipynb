```python
class annual_fee:

    def __init__(self, name):
        self.name = name
        self.anual_cost = []
        self.subject_name = None

    def subject(self):
        self.subject_name = input(
            "Enter subject (HR / Finance / DS / Marketing): "
        )

        if self.subject_name == "DS":
            print("Analytics course not available for DS")
        else:
            print(f"{self.name} is taking {self.subject_name}")

    def add_course_expense(self):
        course = input("Is your course analytics (yes/no): ")
        if course == "yes":
            self.anual_cost.append(200000 + 200000 * 0.1)

    def add_hostel_expense(self):
        hostel = input("Do you want hostel (yes/no): ")
        if hostel == "yes":
            self.anual_cost.append(200000)

    def add_food_expense(self):
        months = int(input("Enter number of months for food: "))
        self.anual_cost.append(months * 2000)

    def add_transport_expense(self):
        sem = int(input("How many semesters for transport: "))
        self.anual_cost.append(sem * 13000)

    def total_annual_expense(self):
        total = sum(self.anual_cost)
        return f"{self.name} total annual cost for {self.subject_name} : {total} ₹"

    # ONE METHOD TO CALL EVERYTHING
    def calculate_fee(self):
        self.subject()
        self.add_course_expense()
        self.add_hostel_expense()
        self.add_food_expense()
        self.add_transport_expense()
        print(self.total_annual_expense())
```

```python
karan=annual_fee("karan")
karan.calculate_fee()
```
