```python
from datetime import datetime

# Interest buckets
# (start_day, end_day, public_rate, senior_rate)

interest_buckets = [
    (7, 45, 5.75, 6.25),
    (46, 179, 6.25, 6.75),

    (180, 210, 6.35, 6.85),
    (211, 364, 6.40, 6.90),
    (365, 729, 7.00, 7.50),
    (730, 1094, 6.75, 7.25),
    (1095, 1824, 6.70, 7.20),
    (1825, 3650, 6.60, 7.10)
]

# Input dates
start_date = input("Enter FD start date (dd-mm-yyyy): ")
end_date = input("Enter FD end date (dd-mm-yyyy): ")

# Convert string to date
s_date = datetime.strptime(start_date, "%d-%m-%Y")
e_date = datetime.strptime(end_date, "%d-%m-%Y")

# Calculate FD duration
days = (e_date - s_date).days
print("FD Duration:", days, "days")

# Customer type
customer_type = input("Enter customer type (public/senior): ").lower()

# Find interest rate
interest_rate = None

for start, end, public_rate, senior_rate in interest_buckets:
    if days >= start and days <= end:
        if customer_type == "public":
            interest_rate = public_rate
        elif customer_type == "senior":
            interest_rate = senior_rate
        break

# Output result
if interest_rate:
    print("Applicable Interest Rate:", interest_rate, "%")
else:
    print("No interest bucket found")
```
